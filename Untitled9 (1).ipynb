```python
data['review'] = data['review'].str.lower()
data['sentiment'] = data['sentiment'].str.lower()
```

```python
import re

data['review'] = data['review'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x)) 

data['review'] = data['review'].apply(lambda x: ' '.join(word for word in x.split() if len(word) > 1)) 
```

```python
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords



# удаление стоп-слов для английского языка
stop_words = set(stopwords.words('english'))
df['review'] = df['review'].apply(lambda x: ' '.join([word for word in x.lower().split() if word not in stop_words]))

# стемминг каждого слова в тексте датасета
ps = PorterStemmer()
df['review'] = df['review'].apply(lambda x: ' '.join([ps.stem(word) for word in x.lower().split()]))

# проверка результата
print(df.head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()  
X = vectorizer.fit_transform(data['review'].values.astype('U'))
y = data['sentiment'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
x_pred = log_reg.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, x_pred)
```
