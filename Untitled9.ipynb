```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('example.csv', on_bad_lines='skip')
```

```python
data['review'] = data['review'].str.lower()
data['sentiment'] = data['sentiment'].str.lower()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['sentiment'] = le.fit_transform(data['sentiment'])
```

```python
data
```

```python
import re

data['review'] = data['review'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x)) 

data['review'] = data['review'].apply(lambda x: ' '.join(word for word in x.split() if len(word) > 1)) 
```

```python
data
```

```python
data['review'] = data['review'].str.replace('br', '')
```

```python
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
import pandas as pd 

vectorizer = CountVectorizer() 
X = vectorizer.fit_transform(data['review']) 

print(X.shape)
data
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Загрузка датасета
df = data

# Удаление стоп-слов для английского языка
stop_words = set(stopwords.words('english'))
df['review'] = df['review'].apply(lambda x: ' '.join([word for word in x.lower().split() if word not in stop_words]))

# Проверка результата
print(df.head())
```

```python
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords



# удаление стоп-слов для английского языка
stop_words = set(stopwords.words('english'))
df['review'] = df['review'].apply(lambda x: ' '.join([word for word in x.lower().split() if word not in stop_words]))

# стемминг каждого слова в тексте датасета
ps = PorterStemmer()
df['review'] = df['review'].apply(lambda x: ' '.join([ps.stem(word) for word in x.lower().split()]))

# проверка результата
print(df.head())
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# загрузка датасета
df = data

# настройки токенизации (удаление пунктуации и цифр)
tokenizer = nltk.RegexpTokenizer(r'\w+')

# токенизация каждой строки в столбце 'text'
df['review'] = df['review'].apply(lambda x: tokenizer.tokenize(x.lower()))

# проверка результата
print(df.head())
```
